```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Create dataset
np.random.seed(42)
X = np.random.rand(1000, 2)
y = (X[:, 0] + X[:, 1] > 1).astype(int)
```

```python
# Step 2: Plot dataset
plt.figure(figsize=(6, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Class 1')

# Add labels and grid
plt.title("Generated Dataset (x1 + x2 > 1 â†’ Class 1)")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 2: Define model structure
def build_model():
    model = Sequential([
        Dense(8, activation='relu', input_dim=2),
        Dense(8, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    return model
```

# Step 3: Model 1 - SGD with Momentum (no NAG)

```python
model1= build_model()
model1.compile(optimizer=SGD(learning_rate=0.1,momentum=0.9,nesterov=False), loss='binary_crossentropy', metrics=['accuracy'])
history = model1.fit(X, y, epochs=50, batch_size=10, verbose=1)
```

```python
model1= build_model()
model1.compile(optimizer=SGD(learning_rate=0.1,momentum=0.9,nesterov=False), loss='binary_crossentropy', metrics=['accuracy'])
history = model1.fit(X, y, epochs=50, batch_size=10, verbose=1)
```

```python
model2= build_model()
model2.compile(optimizer=SGD(learning_rate=0.1,momentum=0.9,nesterov=True),loss="binary_crossentropy",metrics=["accuracy"])
history2=model2.fit(X,y,epochs=50,batch_size=10,verbose=1)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(history.history['accuracy'], label='Momentum (no NAG)', color='blue')
plt.plot(history2.history['accuracy'], label='Nesterov Accelerated Gradient (NAG)', color='red')
plt.title('Training Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
